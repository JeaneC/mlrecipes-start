We will use one example of each type of flower as training data

```python
import numpy as np 
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
test_idx = [0,50,100] #3 DIFFERENT LOCATIONS OF 3 DIFFERENT FLOWERS
# training data
train_target = np.delete(iris.target, test_idx) # the target is the data of everything but the elements at 0,50,100
train_data = np.delete(iris.data, test_idx, axis = 0) # same as above

# testing data
test_target = iris.target[test_idx] # the target here are teh elements with the index of 0,50,100
test_data = iris.data[test_idx]

clf = tree.DecisionTreeClassifier()
clf.fit(train_data, train_target) #Find common features 

print(test_target)
print(clf.predict(test_data))
```

---
Purely just to visualize the data

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydot

dot_data = StringIO()
tree.export_graphviz(clf, out_file=dot_data, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         impurity=False)  
graph = pydot.graph_from_dot_data(dot_data.getvalue())

graph[0].write_png("iris.png")
Image("iris.png")
```

- Start by reading from the top
- Each node asks a question about each feature if it's true go left, if it's false go right
- Samples indicate how much data it is true for?
